```python

def tend(x ,mult,add) : 
    #mult = np.random.uniform(1, 5)
    #add = np.random.uniform(1,10)
    #return (x/(sais+mult) + add)
    return(mult*x + add )
```

```python
def tendhm(x):
    t = -((x-50)/5)**2+100
    d = np.random.uniform(50,100)
    return(t/d + 1)
```

```python
def tendhadd(x):
    t = -((x-50)/5)**2+100
    a = np.random.uniform(4,15)
    b = np.random.uniform(4,10)
    return(t/a + b)
```
