```python
def tendhm(x):
    t = -((x-50)/5)**2+100
    d = np.random.uniform(50,100)
    return(t/d + 1)
```

```python
#création multiples

#normalisation 
df = pd.DataFrame()
liste_ts = []
liste_type_tend = []
liste_sais = []
for j in range (0,10000 ):
    
    sais = np.random.randint(7,24)

    y = []
    #x = []

    #liste tendance
    t = []

    m = np.random.uniform(0.1,1.5)
    n = 1/m
    pi = np.pi
    a = np.random.uniform(1,13)
    d = np.random.uniform(0,2*pi)
    #pour tend
    mult = np.random.uniform(0.05, 0.1)
    add = np.random.uniform(1,10)
    
    l_mult = np.random.uniform(0.8, 1.2, size=(sais))
    
    type_tendance = random.choice(['add','mult'])
    
    
    #print(sais)
    for i in range (0,100 ):
        #x.append(i)
        rand = np.random.uniform(0.93,1.07)
        tendance = tend(i,mult,add)
        #tendanceh = tendhm(i)
        #tendanceha = tendhadd(i)
        #tendance = 1
        #y.append(((((m*np.cos(pi*n*i)+sais/2 +10)/20)*l_mult[i%sais])*rand)*tendance)
        #y.append((((m*np.cos((2*pi*i)/sais)+a)*l_mult[i%sais])*rand)*tendance)

        #t.append(tendance)
        s2 = sais/2
        sig = s2*np.cos(((2*pi*i+(d*sais))/sais))+s2 #val signal 0-sais

        #y.append((m*sig+a)*l_mult[i%sais]*rand*tendance)#tendance mult
        if type_tendance == 'add':
            y.append((m*sig+a)*l_mult[i%sais]*rand+tendance)#tendance add
        else:
            y.append((m*sig+a)*l_mult[i%sais]*rand*tendance)#tendance mult
        #y.append((m*sig+a)*l_mult[i%sais]*rand*tendanceh)#tendance mult
        #y.append((m*sig+a)*l_mult[i%sais]*rand*4+tendanceha)
    
    #normalisation
    a = np.array(y)
    a = (a-a.min())/(a.max()-a.min())
    
    liste_ts.append(a)
    liste_type_tend.append(type_tendance)
    liste_sais.append(sais)
    
df['TS'] = liste_ts
df['type_tendance'] = liste_type_tend
df['freq'] = liste_sais
```
